```python
#Import first data source- LINKS SMALL
links_file = "../ETL_Project_original_Data/links_small.csv"
links_df = pd.read_csv(links_file)

#links_df
#links_df.dtypes
#links_df.columns
```

```python
#Import second data source- RATINGS 
ratings_file = "../ETL_Project_original_Data/ratings_small.csv"
ratings_df = pd.read_csv(ratings_file)

#ratings_df.dtypes
```

```python
#JOIN 1: Link with ratings
## Chose to merge these two tables first because of size and easy to connect with movieID
links_ratings_merge = pd.merge(links_df,ratings_df,on ="movieId")

#links_ratings_merge
#links_ratings_merge.dtypes
#links_ratings_merge.columns
#links_ratings_merge
```
